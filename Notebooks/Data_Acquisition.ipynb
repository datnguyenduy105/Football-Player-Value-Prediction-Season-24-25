```python
import random

# Function to check if a string contains numbers.  Used for market value filtering.
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

# Main scraping function. Takes a list of team links as input.
def scrape_links(links):
    PlayersList = []
    ValuesList = []
    TeamsList = []
    

    # List of user agents to rotate, helping to avoid being blocked.
    user_agent_list = [
            # Desktop Browsers
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:120.0) Gecko/20100101 Firefox/120.0",
            "Mozilla/5.0 (Macintosh; Intel Mac OS X 14_0) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.0 Safari/605.1.15",
            "Mozilla/5.0 (Macintosh; Intel Mac OS X 14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",

            # Mobile Browsers
            "Mozilla/5.0 (iPhone; CPU iPhone OS 17_1_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.1 Mobile/15E148 Safari/604.1",
            "Mozilla/5.0 (Linux; Android 14; Pixel 8 Pro) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Mobile Safari/537.36",
            "Mozilla/5.0 (Linux; Android 13; SAMSUNG SM-G991B) AppleWebKit/537.36 (KHTML, like Gecko) SamsungBrowser/23.0 Chrome/116.0.0.0 Mobile Safari/537.36",
            "Mozilla/5.0 (iPad; CPU OS 17_1_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.0 Mobile/15E148 Safari/604.1",

            # Search Bots
            "Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)",
             "Mozilla/5.0 (compatible; Bingbot/2.0; +http://www.bing.com/bingbot.htm)"
    ]
    
    for link in links:
        team_players = []
        team_values = []
        
        # Extract team name from the URL.        
        start = len("https://www.transfermarkt.us/")
        end = link.find("/startseite")
        team_name = link[start:end]
        team_name = team_name.replace('-', ' ')

        # Randomly select a user agent for this request.
        user_agent = random.choice(user_agent_list)
        headers = {'User-Agent': user_agent}

        # Make the HTTP request to the URL.
        req = urllib.request.Request(url=link, headers=headers)
        with urllib.request.urlopen(req) as response:
            page_html = response.read()

        # Parse the HTML using BeautifulSoup.
        soup = BeautifulSoup(page_html, 'html.parser')
       

        # Extract player names from the table. This section uses a more robust method
        # to find player names compared to the commented-out code.
        inline_tables = soup.find_all("table", {"class": "inline-table"})

        # Extract market values. Handles cases where value is missing ('0').
        for table in inline_tables:
            player_link = table.find("a")
            if player_link:
                player_name = player_link.text.strip()
                if player_name not in team_players:
                    team_players.append(player_name)
                    PlayersList.append(player_name)
                    TeamsList.append(team_name)

        values = soup.find_all("td", {"class": "rechts hauptlink"})
        for i in range(len(values)):
            value = values[i].text
            if hasNumbers(value) == True:
                ValuesList.append(value)
                team_values.append(value)

            else:
                ValuesList.append('0')
                team_values.append('0')

        print("num players: ", len(team_players), "  num values: ", len(team_values))
        
                
    print(len(PlayersList))
    print(len(ValuesList))

    # Create a Pandas DataFrame from the collected data.
    df = pd.DataFrame({'Players': PlayersList, "Values": ValuesList, "Team": TeamsList})
    return df
```

```python
# Scrape data for each league and save to a CSV file.
premier_league_df = scrape_links(premier_league_links)
```
